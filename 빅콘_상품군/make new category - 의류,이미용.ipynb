# 의류(총 12개)

#### - 여성/남성  
#### - 가죽/모피/무스탕, 자켓/코트/패딩, 상의1, 상의2, 하의, 세트상품

```python
df_clothes = df[df['상품군']=='의류']
```

```python
print(df_clothes.shape)
```

```python
df_clothes['gender'] = np.where(df_clothes['상품명'].apply(lambda x:'남성' in x),'남성','여성') 
```

```python
leather = ["가죽","밍크","폭스카라","폭스퍼","모피","무스탕"]
outer = ["자켓","재킷","점퍼","코트","롱패딩","덕다운","패딩 세트","패딩코트세트","구스다운","윈드브레이커"]
top1 = ["니트","맨투맨","터틀넥","모크넥"]
top2 = ["기모티","셔츠","블라우스","베스트"]
bottom = ["데님","팬츠","스커트","원피스","바지"]
sett=["세트","수트","풀코디","앙상블","멀티","액티비티","라인핏"]

df_clothes['subCategory'] = np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in leather)),"가죽/모피/무스탕",\
                            np.where(df_clothes['상품명'].apply(lambda x:'+' in x),"세트상품",\
                            np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in outer)),"자켓/코트/패딩",\
                            np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in top1)),"상의1",\
                            np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in top2)),"상의2",\
                            np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in bottom)),"하의",\
                            np.where(df_clothes['상품명'].apply(lambda x:any(k in x for k in sett)),"세트상품","자켓/코트/패딩")))))))
```

```python
df_clothes['newCategory'] = df_clothes['gender']+df_clothes['subCategory']
```

```python
df_clothes.head()
```

```python
df_clothes=df_clothes.drop(['gender','subCategory'],axis=1)
df_clothes.head()
```
